```python
import pandas as pd
import numpy as np
# Load the dataset
df = pd.read_csv("anime.csv")

# Convert 'episodes' to numeric, replace non-numeric with NaN
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')

# Drop rows with missing values in numeric columns
df_clean = df.dropna(subset=['episodes', 'rating', 'members'])
print(df_clean)
```

```python
#Maximum number of episodes
# Find the maximum number of episodes
max_episodes = np.max(df_clean['episodes'].values)
print("Maximum episodes:", max_episodes)
```

```python
# Sum of all members
total_members = np.sum(df_clean['members'].values)
print("Sum of all members:", total_members)
```

```python
# Count how many unique episode counts are there
unique_episodes = np.unique(df_clean['episodes'].values).shape[0]
print("Unique episode counts:", unique_episodes)
```

```python
# Variance of the rating values
variance_ratings = np.var(df_clean['rating'].values)
print("Variance of ratings:", variance_ratings)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
anime_df = pd.read_csv('/content/anime.csv')

# Check the columns
print(anime_df.columns)

# Start Subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 rows, 2 columns of plots
fig.suptitle('Anime Dataset Analysis', fontsize=16)

# --- Plot 1: Histogram of Ratings ---
axs[0, 0].hist(anime_df['rating'].dropna(), bins=10, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Histogram of Ratings')
axs[0, 0].set_xlabel('Rating')
axs[0, 0].set_ylabel('Count')

# --- Plot 2: Bar plot of Top 5 Genres ---
# If genre column exists
if 'genre' in anime_df.columns:
    top_genres = anime_df['genre'].value_counts().head(5)
    axs[0, 1].bar(top_genres.index, top_genres.values, color='lightgreen', edgecolor='black')
    axs[0, 1].set_title('Top 5 Genres')
    axs[0, 1].set_xlabel('Genre')
    axs[0, 1].set_ylabel('Count')
    axs[0, 1].tick_params(axis='x', rotation=45)


# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # To not overlap with the main title
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("anime.csv")

# Convert episodes to numeric (handle 'Unknown')
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')

# Drop rows with missing values in 'episodes' or 'rating'
df_clean = df.dropna(subset=['episodes', 'rating'])

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['episodes'], df_clean['rating'], color='purple', alpha=0.5)
plt.title('Episodes vs Rating of Anime')
plt.xlabel('Number of Episodes')
plt.ylabel('User Rating')
plt.grid(True)
plt.show()
```
